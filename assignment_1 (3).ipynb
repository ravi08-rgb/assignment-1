```python
# Write a Python program to print "Hello, World
print("Hello,World")
# Write a Python program that displays your name and age
name="Ravi verma"
age=24
 print("Name=",name)
 print("Age=",age)
# Write code to print all the pre-defined keywords in Python using the keyword library8

import keyword

print(keyword.kwlist)
# Write a program that checks if a given word is a Python keyword.

import keyword

def is_keyword(word):
    """Checks if a given word is a Python keyword.

    Args:
        word: The word to check.

    Returns:
        True if the word is a keyword, False otherwise.
    """
    return keyword.iskeyword(word)

# Example usage
word1 = "for"
word2 = "hello"
print(f"'{word1}' is a keyword: {is_keyword(word1)}")
print(f"'{word2}' is a keyword: {is_keyword(word2)}")

#5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
a = [1, 2, 3]
b = ['apple', 'orange', 'cherry']

res = list(zip(a, b))

print(res)
6. Write a function to demonstrate the behavior of mutable and immutable arguments
def modify_arguments(immutable_arg, mutable_arg):
    """
    Demonstrates the behavior of mutable and immutable arguments.

    Args:
      immutable_arg: An immutable object (e.g., integer, string, tuple).
      mutable_arg: A mutable object (e.g., list, dictionary).
    """
    # Attempt to modify the immutable argument
    immutable_arg = immutable_arg + 1  # Creates a new object
    print(f"Inside function - Immutable argument: {immutable_arg}")

    # Modify the mutable argument
    mutable_arg.append(4) # Modifies the original object
    print(f"Inside function - Mutable argument: {mutable_arg}")

# Example usage
immutable_value = 5
mutable_value = [1, 2, 3]

print(f"Before function call - Immutable value: {immutable_value}")
print(f"Before function call - Mutable value: {mutable_value}")

modify_arguments(immutable_value, mutable_value)

print(f"After function call - Immutable value: {immutable_value}")
print(f"After function call - Mutable value: {mutable_value}")


7. Write a program that performs basic arithmetic operations on two user-input numbers.

def perform_arithmetic_operations():
    """
    Prompts the user for two numbers and performs basic arithmetic operations.
    """
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        # Perform basic operations
        addition = num1 + num2
        subtraction = num1 - num2
        multiplication = num1 * num2
        division = num1 / num2

        # Display the results
        print("Addition:", addition)
        print("Subtraction:", subtraction)
        print("Multiplication:", multiplication)
        print("Division:", division)

    except ValueError:
        print("Invalid input. Please enter numbers.")

# Call the function to run the program
perform_arithmetic_operations(


8. Write a program to demonstrate the use of logical operators.

#include <iostream>

using namespace std;

int main() {
    int num1, num2;

    // Get input from the user
    cout << "Enter two integers: ";
    cin >> num1 >> num2;

    // Demonstrate logical operators (AND, OR, NOT)
    cout << "\nDemonstrating logical operators:" << endl;

    // AND (&&) - Both operands must be true for the entire expression to be true
    if (num1 > 0 && num2 > 0) {
        cout << "Both numbers are positive." << endl;
    } else {
        cout << "At least one number is not positive." << endl;
    }

    // OR (||) - At least one operand must be true for the entire expression to be true
    if (num1 > 0 || num2 > 0) {
        cout << "At least one number is positive." << endl;
    } else {
        cout << "Neither number is positive." << endl;
    }

    // NOT (!) - Inverts the truth value of an expression
    if (!(num1 == 0)) { // Check if num1 is not equal to 0
        cout << "Number 1 is not zero." << endl;
    } else {
        cout << "Number 1 is zero." << endl;
    }

    return 0;
}

9. Write a Python program to convert user input from string to integer, float, and boolean types.
  def convert_input():
    string_input = input("Enter a string: ")

    try:
        int_input = int(string_input)
        print(f"Integer: {int_input}")
    except ValueError:
        print("Not a valid integer.")

    try:
        float_input = float(string_input)
        print(f"Float: {float_input}")
    except ValueError:
        print("Not a valid float.")

    if string_input.lower() == 'true':
        bool_input = True
    elif string_input.lower() == 'false':
        bool_input = False
    else:
        print("Not a valid boolean.")
        return

    print(f"Boolean: {bool_input}")

convert_input()
10. Write code to demonstrate type casting with list elements
                             # Initial list with mixed data types
my_list = ['10', '20', 30, 40.5, '50']
print("Original list:", my_list)

# Type casting elements to integers
int_list = [int(x) for x in my_list]
print("List casted to integers:", int_list)

# Type casting elements to floats
float_list = [float(x) for x in my_list]
print("List casted to floats:", float_list)

# Type casting elements to strings
str_list = [str(x) for x in my_list]
print("List casted to strings:", str_list)

# Type casting elements back to their original type
mixed_list = [int(x) if isinstance(x, str) and x.isdigit() else float(x) if isinstance(x, str) and '.' in x else x for x in my_list]
print("List casted to mixed types:", mixed_list)

11. Write a program that checks if a number is positive, negative, or zero.
                             def check_number(num):
  """
  Checks if a number is positive, negative, or zero.

  Args:
    num: The number to check.

  Returns:
    A string indicating whether the number is positive, negative, or zero.
  """
  if num > 0:
    return "Positive"
  elif num < 0:
    return "Negative"
  else:
    return "Zero"

# Example usage:
number = int(input("Enter a number: "))
result = check_number(number)
print(f"The number is {result}.")
 12. Write a for loop to print numbers from 1 to 10.
    for i in range(1, 11):
    print(i)
13 Write a Python program to find the sum of all even numbers between 1 and 50
    total = 0
for number in range(2, 51, 2):
    total += number
print(total)
14. Write a program to reverse a string using a while loop.
  def reverse_string(s):
    reversed_s = ""
    length = len(s) - 1
    while length >= 0:
        reversed_s += s[length]
        length -= 1
    return reversed_s

input_string = "hello"
reversed_string = reverse_string(input_string)
print(reversed_string)
15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

```
